In total we have 3948 matches. It is often said that there is no way to know who wins a hockey game, because there are so many lucky bounces, goalies are brickwalls, the best team has a bad day etc. It's not uncommon that the worst team in the league beats the best team in the league 1-0, even though the shots were 12-43. To quickly check that none of these match stats correlate with who won the game, I will create some new columns: 
- `home_win` - True if `home_goals` are more than `away_goals`, else False
- `shot_diff` - The difference between `home_shots` and `away_shots`
- `sog_diff` - The difference between `home_sog` and `away_sog`
- `pp_diff` - The difference between `home_pp` and `away_pp`
- `ppp_diff` - The difference between `home_ppp` and `away_ppp`

```python
all_matches['home_win'] = all_matches['home_goals'] > all_matches['away_goals']
all_matches['shot_diff'] = all_matches['home_shots'] - all_matches['away_shots']
all_matches['sog_diff'] = all_matches['home_sog'] - all_matches['away_sog']
all_matches['pp_diff'] = all_matches['home_pp'] - all_matches['away_pp']
all_matches['ppp_diff'] = all_matches['home_ppp'] - all_matches['away_ppp']

all_matches.head()
```

From the heatmap above we can see that none of the match stats correlate much with `home_win`, meaning that the stats don't provide valuable information about who will win the game. All the correlations are close to 0, meaning that there is no correlation, except `ppp_diff` has a correlation of 0.34, but using only that one variable in our machine learning model probably won't get us far.

## Using Stat Trends To Predict the Winners

The idea I have for the machine learning model is to create new features that use the "previous 5 games" trends for all the matches. I think that trends or "momentum" have a huge factor in who wins the game. For example in the NHL last year Edmonton lost almost all the games when the season started, but after that they got a good momentum and ended up in the finals.

### Previous 5 Games

I will create the following indicators for each team:
- `prev_5_win_pct` - The percentage of how many games the team won in the last 5 games
- `prev_5_sog_diff` - The sum of shots for - the sum of shots against
- `prev_5_shot_pct_avg` - The average of the difference between shot pct for and shot pct against
- `prev_5_points_avg` - The amount of points in the last 5 games divided by 5
- `prev_5_pp_pct_avg` - The average amount of powerplays in the last 5 games
- `prev_5_pk_pct_avg` - The average amount of power kills in the last 5 games
- `prev_5_goal_diff` - The sum of goals for - the sum of goals against
- `prev_5_save_pct` - The average save percentage in the last 5 games
- `prev_5_penalties_taken_avg` - The average amount of penalties taken in the last 5 games
- `prev_5_goals_scored_avg` - The average amount of goals scored in the last 5 games
- `prev_5_goals_against_avg` - The average amount of goals against in the last 5 games

I think the easiest way to do this is to use SQL to create this trending data for each team, and then later combine the matches with the trending data. I will start with the team Blues to create the first table, and then I will assign it to all the teams. First I will need to drop the columns created earlier, because otherwise the SQL table will have columns with the same name.

```python
%%sql

WITH blues AS (
    SELECT *,
        -- win counter
        CASE
           WHEN home_team = "Blues" AND home_goals > away_goals THEN True
           WHEN away_team = "Blues" AND away_goals > home_goals THEN True
           ELSE False
        END AS win,
        -- regular time win counter
        CASE
           WHEN home_team = "Blues" AND home_goals > away_goals AND ot=0 THEN True
           WHEN away_team = "Blues" AND away_goals > home_goals AND ot=0 THEN True
           ELSE False
        END AS reg_time_win,
        -- ot wins
        CASE
            WHEN home_team = "Blues" AND home_goals > away_goals AND ot=1 THEN True
            WHEN away_team = "Blues" AND away_goals > home_goals AND ot=1 THEN True           
           ELSE False
        END AS ot_win,
        -- ot losses
        CASE
            WHEN home_team = "Blues" AND home_goals < away_goals AND ot=1 THEN True
            WHEN away_team = "Blues" AND away_goals < home_goals AND ot=1 THEN True           
           ELSE False
        END AS ot_loss,
        -- regular time losses
        CASE
            WHEN home_team = "Blues" AND home_goals < away_goals AND ot=0 THEN True
            WHEN away_team = "Blues" AND away_goals < home_goals AND ot=0 THEN True           
           ELSE False
        END AS loss,
        -- shots on goal difference
        CASE
            WHEN home_team = "Blues" THEN home_sog-away_sog
            WHEN away_team = "Blues" THEN away_sog-home_sog
        END AS sog_diff,
        -- shot percent difference
        CASE
            WHEN home_team = "Blues" THEN home_sog_pct-away_sog_pct
            WHEN away_team = "Blues" THEN away_sog_pct-home_sog_pct
        END AS shot_pct_diff,
        -- points
        3 * (CASE WHEN home_team = "Blues" AND home_goals > away_goals AND ot = 0 THEN 1
                  WHEN away_team = "Blues" AND away_goals > home_goals AND ot = 0 THEN 1
                  ELSE 0
             END) +
        2 * (CASE WHEN home_team = "Blues" AND home_goals > away_goals AND ot = 1 THEN 1
                  WHEN away_team = "Blues" AND away_goals > home_goals AND ot = 1 THEN 1
                  ELSE 0
             END) +
        1 * (CASE WHEN home_team = "Blues" AND home_goals < away_goals AND ot = 1 THEN 1
                  WHEN away_team = "Blues" AND away_goals < home_goals AND ot = 1 THEN 1
                  ELSE 0
             END) +
        0 * (CASE WHEN home_team = "Blues" AND home_goals < away_goals AND ot = 0 THEN 1
                  WHEN away_team = "Blues" AND away_goals < home_goals AND ot = 0 THEN 1
                  ELSE 0
             END) AS points,
        -- powerplay pct
        CASE 
            WHEN home_team = "Blues" THEN home_ppp
            WHEN away_team = "Blues" THEN away_ppp
        END AS pp_pct,
        --penalty kill pct 
        CASE 
            WHEN home_team = "Blues" THEN 1-(away_ppg/away_pp)
            WHEN away_team = "Blues" THEN 1-(home_ppp/home_pp)
        END AS pk_pct,
        -- goal differential
        CASE 
            WHEN home_team = "Blues" THEN home_goals-away_goals
            WHEN away_team = "Blues" THEN away_goals-home_goals
        END as goal_diff,
        -- save_pct
        CASE 
            WHEN home_team = "Blues" THEN 1-(1.0*away_goals/away_sog)
            WHEN away_team = "Blues" THEN 1-(1.0*home_goals/home_sog)
        END as save_pct,
        -- Penalties taken
        CASE 
            WHEN home_team = "Blues" THEN away_pp
            WHEN away_team = "Blues" THEN home_pp
        END as penalties_taken,
        -- Goals scored
        CASE 
            WHEN home_team = "Blues" THEN home_goals
            WHEN away_team = "Blues" THEN away_goals
        END as goals_scored,
        -- Goals against
        CASE 
            WHEN home_team = "Blues" THEN away_goals
            WHEN away_team = "Blues" THEN home_goals
        END as goals_against
        FROM all_matches
     WHERE (home_team = "Blues" OR away_team = "Blues") AND season = '2015-2016'
)

SELECT  "Blues" AS team_name,
        date,
        ROW_NUMBER() OVER(ORDER BY date) AS games_played,
        season,
        home_team,
        home_goals,
        away_team,
        away_goals,
        --prev 5 wins divided by 5
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        0.2*SUM(win) OVER(ORDER BY date
                          ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_win_pct,
        -- sum of sog difference
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        SUM(sog_diff) OVER(ORDER BY date
                            ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_sog_diff,
        -- avg shot pct difference    
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        AVG(shot_pct_diff) OVER(ORDER BY date
                                ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_shot_pct_diff,
        -- avg points (win-loss-draw-ratio)
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        AVG(points) OVER(ORDER BY date
                         ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_points_avg,
        -- Average power play percentage
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        AVG(pp_pct) OVER(ORDER BY date
                         ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_pp_pct,
        -- Average penalty kill percentage
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        AVG(pk_pct) OVER(ORDER BY date
                         ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_pk_pct,
        -- Goal differential
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        AVG(goal_diff) OVER(ORDER BY date
                            ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_goal_diff,
        -- save pct differential
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        AVG(save_pct) OVER(ORDER BY date
                           ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_save_pct_avg,
        -- Average penalties taken
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        AVG(penalties_taken) OVER(ORDER BY date
                                  ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_penalties_taken_avg,
        -- Average goals scored
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        AVG(goals_scored) OVER(ORDER BY date
                                ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_goals_scored_avg,
        -- Average goals allowed
        CASE WHEN COUNT(*) OVER(ORDER BY date) > 5 THEN
        AVG(goals_against) OVER(ORDER BY date
                                ROWS between 5 PRECEDING AND 1 PRECEDING) END AS prev_5_goals_against_avg
  FROM blues
```

## Combining Matches and Trending Data

To combine the matches with the trending data, we can simply merge them when `date`s and `season`s match and when `team_name` matches with `home_team` or `away_team`. We also don't need any of the original match data, so we can clean that.

```python
all_matches_cleaned = all_matches[['date', 'home_team', 'away_team', 'season']]
all_matches_cleaned = all_matches_cleaned.copy() 
```

```python
#fixing the dates
all_matches_cleaned['date'] = pd.to_datetime(all_matches_cleaned['date'])
all_teams_trends['date'] = pd.to_datetime(all_teams_trends['date'])

home_merge = pd.merge(all_matches_cleaned, all_teams_trends, left_on=['date', 'season', 'home_team'],
                      right_on=['date', 'season', 'team_name'], how='left')
away_merge = pd.merge(all_matches_cleaned, all_teams_trends, left_on=['date', 'season', 'away_team'],
                      right_on=['date', 'season', 'team_name'], how='left')
#adding suffixes so we can use concat
home_merge = home_merge.add_suffix('_home')
away_merge = away_merge.add_suffix('_away')

merged_matches = pd.concat([home_merge, away_merge], axis=1)
```

There are multiple columns that are useless for machine learning. For example team names are now multiple times, dates are multiple times, and there is no reason to give the machine learning model prev_x_home and prev_x_away, when they provide the same information if we just take the result of prev_x_home - prev_x_away to get a final feature of prev_x_home_adv, which tells us the positive or negative value the home team has in that stat. So now I will remove all the useless columns and create the final feature columns.

```python
merged_matches['prev_5_win_pct_home_adv'] = merged_matches['prev_5_win_pct_home'] - merged_matches['prev_5_win_pct_away']
merged_matches['prev_5_win_pct_home_adv'] = merged_matches['prev_5_win_pct_home'] - merged_matches['prev_5_win_pct_away']
merged_matches['prev_5_sog_diff_home_adv'] = merged_matches['prev_5_sog_diff_home'] - merged_matches['prev_5_sog_diff_away']
merged_matches['prev_5_shot_pct_diff_home_adv'] = merged_matches['prev_5_shot_pct_diff_home'] - merged_matches['prev_5_shot_pct_diff_away']
merged_matches['prev_5_points_avg_home_adv'] = merged_matches['prev_5_points_avg_home'] - merged_matches['prev_5_points_avg_away']
merged_matches['prev_5_pp_pct_home_adv'] = merged_matches['prev_5_pp_pct_home'] - merged_matches['prev_5_pp_pct_away']
merged_matches['prev_5_pk_pct_home_adv'] = merged_matches['prev_5_pk_pct_home'] - merged_matches['prev_5_pk_pct_away']
merged_matches['prev_5_goal_diff_home_adv'] = merged_matches['prev_5_goal_diff_home'] = merged_matches['prev_5_goal_diff_away']
merged_matches['prev_5_save_pct_avg_home_adv'] = merged_matches['prev_5_save_pct_avg_home'] - merged_matches['prev_5_save_pct_avg_away']
merged_matches['prev_5_penalties_taken_avg_home_adv'] = merged_matches['prev_5_penalties_taken_avg_home'] - merged_matches['prev_5_penalties_taken_avg_away']
merged_matches['prev_5_goals_scored_avg_home_adv'] = merged_matches['prev_5_goals_scored_avg_home'] - merged_matches['prev_5_goals_scored_avg_away']
merged_matches['prev_5_goals_against_avg_home_adv'] = merged_matches['prev_5_goals_against_avg_home'] - merged_matches['prev_5_goals_against_avg_away']

#dropping all the "prev" columns
merged_matches = merged_matches.drop(columns = ['prev_5_win_pct_home', 'prev_5_sog_diff_home',
       'prev_5_shot_pct_diff_home', 'prev_5_points_avg_home',
       'prev_5_pp_pct_home', 'prev_5_pk_pct_home', 'prev_5_goal_diff_home',
       'prev_5_save_pct_avg_home', 'prev_5_penalties_taken_avg_home',
       'prev_5_goals_scored_avg_home', 'prev_5_goals_against_avg_home',
        'prev_5_win_pct_away', 'prev_5_sog_diff_away',
       'prev_5_shot_pct_diff_away', 'prev_5_points_avg_away',
       'prev_5_pp_pct_away', 'prev_5_pk_pct_away', 'prev_5_goal_diff_away',
       'prev_5_save_pct_avg_away', 'prev_5_penalties_taken_avg_away',
       'prev_5_goals_scored_avg_away', 'prev_5_goals_against_avg_away'])
merged_matches.sample(5)
```

## Using SQL to add more features

Now that we know all the features we will use, we have to create the seasonal standings for each date the team's have played a game for the season. This needs to be created after each match to prevent using future data: For example when Blues played their 4th game they had 5 points, not the 59 from the end of the season. This can be done with the matches data. I will again use SQL to do this, and first create one for Blues, and then later loop it for all the teams and seasons.

To calculate the stats we need to use all the previous matches, not just current ones. For example we need to calculate the save percentage using all the previous games shot on goal against and saves made, not just the average results. (For example if goalie saves 50% and 100%, the average would be 75%, but if the first game was 2 shots 1 goal, and the second game was 40 shots with 0 goals, the correct save percentage is 41/42 = 97.6%, not 75%).

```python
%reload_ext sql
%sql sqlite:///data/matches_combined.db
```

```python
%%sql
WITH Blues AS (
    SELECT  date,
            home_team,
            away_team,
            home_goals,
            away_goals,
            OT,
            (60 - (2 * away_pp)) AS home_team_5v5_time,
            (60 - (2 * home_pp)) AS away_team_5v5_time,
            CASE
                WHEN OT = False THEN (home_goals-home_ppg)
                WHEN OT = True AND home_goals > away_goals THEN (home_goals-home_ppg-1)
                WHEN OT = True AND away_goals > home_goals THEN (home_goals-home_ppg)
            END AS home_goals_5v5,
            CASE
                WHEN OT = False THEN (away_goals-away_ppg)
                WHEN OT = True AND away_goals > home_goals THEN (away_goals-away_ppg-1)
                WHEN OT = True AND home_goals > away_goals THEN (away_goals-away_ppg)
            END AS away_goals_5v5,
            home_sog,
            away_sog,
            (away_sog-away_goals) AS home_saves,
            (home_sog-home_goals) AS away_saves,
            home_pp,
            away_pp,
            home_ppg,
            away_ppg,season
            
      FROM all_matches
     WHERE (home_team = 'Blues' OR away_team = 'Blues') 
       AND season = '2015-2016'
),
team_specific_data AS (
    SELECT  date,
            "Blues" as team_name,
            CASE
                WHEN home_team = "Blues" THEN home_team_5v5_time
                WHEN away_team = "Blues" THEN away_team_5v5_time
            END AS time_5v5_one_game,
            CASE
                WHEN home_team = "Blues" THEN away_goals_5v5
                WHEN away_team = "Blues" THEN home_goals_5v5
            END AS goals_against_5v5_one_game,
            CASE
                WHEN home_team = "Blues" THEN home_sog
                WHEN away_team = "Blues" THEN away_sog
            END AS sog_for_one_game,
            CASE
                WHEN home_team = "Blues" THEN home_goals
                WHEN away_team = "Blues" THEN away_goals
            END AS goals_scored_one_game,
            CASE
                WHEN home_team = "Blues" THEN away_sog
                WHEN away_team = "Blues" THEN home_sog
            END AS sog_against_one_game,
            CASE
                WHEN home_team = "Blues" THEN home_saves
                WHEN away_team = "Blues" THEN away_saves
            END AS saves_one_game,
            CASE
                WHEN home_team = "Blues" THEN home_sog-away_sog
                WHEN away_team = "Blues" THEN away_sog-home_sog
            END AS sog_diff_one_game,
            CASE
                WHEN home_team = "Blues" THEN home_pp
                WHEN away_team = "Blues" THEN away_pp
            END AS pp_count_one_game,
            CASE
                WHEN home_team = "Blues" THEN home_ppg
                WHEN away_team = "Blues" THEN away_ppg
            END AS pp_goals_one_game,
            CASE
                WHEN home_team = "Blues" THEN home_goals_5v5
                WHEN away_team = "Blues" THEN away_goals_5v5
            END AS goals_scored_5v5_one_game,
            CASE
                WHEN home_team = "Blues" THEN away_ppg
                WHEN away_team = "Blues" THEN home_ppg
            END AS pk_goals_against_one_game,
            CASE
                WHEN home_team = "Blues" THEN away_pp
                WHEN away_team = "Blues" THEN home_pp
            END AS pk_count_one_game,
            season
            FROM Blues),

--will make one more sum table here for clearer formulas
sum_table AS (
    SELECT  date,
            team_name,
            SUM(time_5v5_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_time_5v5,
            SUM(goals_against_5v5_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_goals_against_5v5,
            SUM(sog_for_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_sog_for,
            SUM(goals_scored_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_goals_scored,
            SUM(sog_against_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_sog_against,
            SUM(saves_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_saves,
            SUM(pp_count_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_pp_count,
            SUM(pp_goals_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_pp_goals_for,
            SUM(goals_scored_5v5_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_goals_scored_5v5,
            SUM(pk_goals_against_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_pk_goals_against,
            SUM(pk_count_one_game) OVER (ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND 1 PRECEDING) AS total_pk_count,
            season
      FROM team_specific_data)
            
--final table, calculates the features
SELECT  date,
        team_name,
        1.0*total_time_5v5/total_goals_against_5v5 AS time_pga_5v5,
        --shot pct + save pct
        (1.0*total_goals_scored/total_sog_for) + (1.0*total_saves/total_sog_against) AS sum_shots_scored_pct_saves,
        total_sog_for - total_sog_against AS sog_diff,
        1.0*total_pp_goals_for/total_pp_count AS pp_pct,
        1.0*total_time_5v5/total_goals_scored_5v5 AS time_pgs_5v5,
        total_pk_goals_against AS pk_ga,
        total_pk_count AS pk_count,
        season
  FROM sum_table
```

### Cleaning the Columns

Now the data is almost ready, but again we need to clean the features, since they are there twice now. So again I will calculate the `home_adv`, meaning that i will take the home team´s data and subtract the away team's data. I will also delete all the prev 5 columns that have `_away` after them.

```python
all_feature_matches['time_pga_5v5_home_adv'] = all_feature_matches['time_pga_5v5'] - all_feature_matches['time_pga_5v5_away']
all_feature_matches['sum_shots_scored_pct_saves_home_adv'] = all_feature_matches['sum_shots_scored_pct_saves'] - all_feature_matches['sum_shots_scored_pct_saves_away']
all_feature_matches['sog_diff_home_adv'] = all_feature_matches['sog_diff'] - all_feature_matches['sog_diff_away']
all_feature_matches['pp_pct_home_adv'] = all_feature_matches['pp_pct']-all_feature_matches['pp_pct_away']
all_feature_matches['time_pgs_5v5_home_adv'] = all_feature_matches['time_pgs_5v5'] - all_feature_matches['time_pgs_5v5_away']
all_feature_matches['pk_ga_home_adv'] = all_feature_matches['pk_ga'] - all_feature_matches['pk_ga_away']
all_feature_matches['pk_count_home_adv'] = all_feature_matches['pk_count'] - all_feature_matches['pk_count_away']

all_feature_matches = all_feature_matches.drop(columns = ['date_away',
       'season_away', 'home_team_away', 'home_goals_away', 'away_team_away',
       'away_goals_away', 'prev_5_win_pct_home_adv_away',
       'prev_5_sog_diff_home_adv_away', 'prev_5_shot_pct_diff_home_adv_away',
       'prev_5_points_avg_home_adv_away', 'prev_5_pp_pct_home_adv_away',
       'prev_5_pk_pct_home_adv_away', 'prev_5_goal_diff_home_adv_away',
       'prev_5_save_pct_avg_home_adv_away',
       'prev_5_penalties_taken_avg_home_adv_away',
       'prev_5_goals_scored_avg_home_adv_away',
       'prev_5_goals_against_avg_home_adv_away', 'home_win_away', 'team_name', 'team_name_away',
        'time_pga_5v5', 'sum_shots_scored_pct_saves', 'sog_diff',
       'pp_pct', 'time_pgs_5v5', 'pk_ga', 'pk_count', 'time_pga_5v5_away',
       'sum_shots_scored_pct_saves_away', 'sog_diff_away', 'pp_pct_away',
       'time_pgs_5v5_away', 'pk_ga_away', 'pk_count_away'])

all_feature_matches.columns
```
